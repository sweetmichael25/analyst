## DATA VISUALIZATION

```python
#import library

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
```

## LINE PLOT

```python
#sample data
x=[1,2,3,4,5]
y=[10,12,5,20,18]
```

```python
#create a line plot
plt.plot(x,y)
plt.xlabel('X-quantity')
plt.ylabel('Y-price')
plt.title('Line Plot')
plt.show()
```

```python
## scattered plot
plt.scatter(x,y)
plt.xlabel('X-quantity')
plt.ylabel('Y-price')
plt.title('Scatter Plot')
plt.show()
```

```python
#BAR PLOT
categories = ['A','B','C','D','E']
values=[25,50,30,40,15]

plt.bar(categories, values)
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar Plot')
plt.show()
```

```python
#to change plot color
categories = ['A','B','C','D','E']
values=[25,50,30,40,15]

plt.bar(categories, values,color ='red')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar Plot')
plt.show()
```

```python
#histogram
data = [1,2,2,3,3,3,4,4,5,5,5,5]

plt.hist(data, bins=5,edgecolor ='red')
plt.ylabel('categories')
plt.xlabel('values')
plt.title('Histogram')
plt.show()
```

```python
# to change histogram color
data = [1,2,2,3,3,3,4,4,5,5,5,5]

plt.hist(data, bins=5,edgecolor ='red',color='yellow')
plt.ylabel('categories')
plt.xlabel('values')
plt.title('Histogram')
plt.show()
```

```python
# pie chart
#load the iris dataset
iris = sns.load_dataset("iris")

# count each species
species_counts = iris['species'].value_counts()

#basic pie chart
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.show()
```

```python
#TO CHANGE COLOR

iris = sns.load_dataset("iris")

# count each species
species_counts = iris['species'].value_counts()

#basic pie chart
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', colors='gyb')
plt.show()
```

```python
#sample data (use your dataset)
data=sns.load_dataset("iris")
#create a pair plot
sns.pairplot(data,hue="species")
plt.show()
```
